```python
import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np

df = pd.read_csv("cleaned-traffic-detectors.csv")
df2 = pd.read_csv("speed_segments_info.csv")

# Create a scatter plot
plt.figure(figsize=(18, 10))
plt.scatter(df['Longitude'], df['Latitude'])

# Add labels and title
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Locations of Traffic Detectors')

# Show the plot
plt.show()
```

```python
df['District'].value_counts().plot(kind='bar')
plt.title('Number of Traffic Detectors in Each District')
plt.xlabel('District')
plt.ylabel('Number of Traffic Detectors')
plt.show()
```

```python
df2['irn_id'].plot(kind='hist', rwidth=0.8)
plt.title('Distribution of IRN ID')
plt.xlabel('IRN ID')
plt.show()
```

```python
df2['ucase(route)'].value_counts().plot(kind='bar')
plt.title('Count of Each Route')
plt.xlabel('Route')
plt.ylabel('Count')
plt.show()
```
