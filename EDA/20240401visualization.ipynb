```python
import pandas as pd
import matplotlib.pyplot as plt
import glob

files = ["traffic_20240401_hour_0.csv", "traffic_20240401_hour_1.csv", "traffic_20240401_hour_2.csv", 
         "traffic_20240401_hour_3.csv", "traffic_20240401_hour_4.csv", "traffic_20240401_hour_5.csv", 
         "traffic_20240401_hour_6.csv", "traffic_20240401_hour_7.csv", "traffic_20240401_hour_8.csv", 
         "traffic_20240401_hour_9.csv", "traffic_20240401_hour_10.csv", "traffic_20240401_hour_11.csv", 
         "traffic_20240401_hour_12.csv", "traffic_20240401_hour_13.csv", "traffic_20240401_hour_14.csv", 
         "traffic_20240401_hour_15.csv", "traffic_20240401_hour_16.csv", "traffic_20240401_hour_17.csv", 
         "traffic_20240401_hour_18.csv", "traffic_20240401_hour_19.csv", "traffic_20240401_hour_20.csv", 
         "traffic_20240401_hour_21.csv", "traffic_20240401_hour_22.csv", "traffic_20240401_hour_23.csv"]
dataframes = [pd.read_csv(file) for file in files]

avg_speed = [df[df["laneType"] == "Slow Lane"]["speed"].mean() for df in dataframes]
avg_volume = [df[df["laneType"] == "Slow Lane"]["totalVolume"].mean() for df in dataframes]
avg_occupancy = [df[df["laneType"] == "Slow Lane"]["totalOccupancy"].mean() for df in dataframes] 

fig, ax = plt.subplots(3, 1, figsize=(10, 6))

# Plot for speed
ax[0].plot(avg_speed, marker='o')
ax[0].set_xlabel('Hour')
ax[0].set_ylabel('Average Speed')
ax[0].set_title('Average Speed of Slow Lane')

# Plot for volume
ax[1].plot(avg_volume, marker='o')
ax[1].set_xlabel('Hour')
ax[1].set_ylabel('Average Volume')
ax[1].set_title('Average Volume of Slow Lane')

# Plot for occupancy
ax[2].plot(avg_occupancy, marker='o')
ax[2].set_xlabel('Hour')
ax[2].set_ylabel('Average Occupancy')
ax[2].set_title('Average Occupancy of Slow Lane')

plt.tight_layout()
plt.show()
```
