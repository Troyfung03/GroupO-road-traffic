Clean the traffic detector data.

```python
import pandas as pd

# Assuming df is your DataFrame
df = pd.read_csv('traffic_detectors.csv')  # replace 'your_file.csv' with your actual file

# List of columns to drop
columns_to_drop = ['Road_TC', 'Road_SC', 'Easting', 'Northing', 'Direction', 'Rotation']

# Drop the unnecessary columns
df = df.drop(columns=columns_to_drop)

# Clean the data
df = df.replace('Central & Western', 'Central and Western')

# Output the cleaned DataFrame as a new CSV file
df.to_csv('cleaned-traffic-detectors.csv', index=False)
```

plot the distribution of the number of traffic detectors in each district

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your data
data = pd.read_csv('cleaned-traffic-detectors.csv')

# Count the number of detectors in each district and sort
district_data = data['District'].value_counts().sort_values(ascending=False).reset_index()
district_data.columns = ['District', 'num_detectors']

# Set the style of the plot
sns.set_theme(style="whitegrid")

# Create the plot
plt.figure(figsize=(10,6))
sns.barplot(x='num_detectors', y='District', data=district_data, palette='viridis')
plt.xlabel('Number of traffic detectors')
plt.ylabel('District')
plt.title('Distribution of traffic detectors in each district')
plt.tight_layout()
plt.show()
```
